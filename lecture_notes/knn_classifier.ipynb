```python
from sklearn.datasets import make_moons

X,y = make_moons(n_samples=300,noise=0.3,random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
```

```python
import numpy as np

# 3) Create a grid over feature space for decision boundary
xx, yy = np.meshgrid(
    np.linspace(-3, +3, 400),
    np.linspace(-3, +3, 400)
)
grid = np.c_[xx.ravel(), yy.ravel()]
pred_grid = knn.predict(grid).reshape(xx.shape)


# 4) Plot decision boundary + points
plt.figure(figsize=(7, 6))

# Decision boundary (background)
plt.contourf(xx, yy, pred_grid, levels=1)

# Training points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=25, edgecolor='k', alpha=0.9, label='train')#, cmap=plt.cm.coolwarm)

# Test points (optional, to visualize)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=25, marker='x', label='test')#,cmap=plt.cm.coolwarm)


plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```
